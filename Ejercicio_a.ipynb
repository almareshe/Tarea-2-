# Gráficas del problema de Dirichlet

**Ejercicio a)**

La solución de la ecuación de Laplace
$$\frac{\partial u^2}{\partial x^2} = \frac{\partial^2 u}{\partial y^2} \;\;\; 0<x<1; \;\; 0<y<1$$
sujeta a las condiciones de frontera:
$$u(0,y)=u(1,y)=0$$
Y la condición inicial:
$$u(x,0)=100,\;\;\; u(x,1)=200$$
Esta dada en términos de la serie de Fourier por la expresión:
$$u(x,y)=\sum_{n=1}^{\infty}sin(nπx)\left[200 \left(\frac{1-(-1^n)}{nπ}\right)cosh(nπy)+sinh(nπy)\left[\frac{1-(-1^n)}{nπ}⋅200 \left(\frac{2-cosh(nπ)}{sinh(nπ)}\right)\right]\right]$$

Este problema modela la transferencia de calor en una barra ideal delgada, con $k=1$.

```python
from pylab import *
```

Primero graficamos la condición inicial.

```python
def recta(x):
    y = zeros(len(x))
    for i in range(len(x)):
      if 0<=x[i]<=(1):
        y[i]=100
    return y
def recta1(x):
    y = zeros(len(x))
    for i in range(len(x)):
      if 0<=x[i]<=(2):
        y[i]=200
    return y

x = linspace(0,1,1000); x1=linspace(1,2,1000)
f = recta(x)
f1= recta1(x)
plot(x,f,label='$u(x,0)=100$'); plot(x1,f1,label='$u(x,1)=200$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$'); 
plt.title("Condición inicial (u(x,0)=100)")
xlim([-0.01,2.01]),ylim([-0.01,215.01]),grid(),legend()
show()
```

Visualizamos la solución completa

```python
def u_xy(x,y,N=30):
    f=zeros((len(x),len(y)))
    for i in range(len(y)):
      for n in range (1,N):
        a=200*(1-(-1)**n)/(n*pi)
        b=a*(2-cosh(n*pi))/sinh(n*pi)
        f[:,i] += sin(n*pi*x)*((a*cosh(n*pi*y[i]))+(b*sinh(n*pi*y[i]))) #No se porque el i con el cosh y el sen h que dependen de y, pero en la ec de calor del profe la tiene con t 
    return f
```

```python
x = linspace(0,1,1000)
y = linspace(0,1,1000)
U = u_xy(x,y)
```

```python
plot(x,U[:,0],label='$u(x,0)$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$'); 
plt.title("Aproximación de la condición inicial")
xlim([-0.01,1.01]),ylim([-0.01,120.5]),grid(),legend()
show()
```

Para diferentes tiempos, lo cual nos permite ver cómo la solución evoluciona cuando $y \to \infty$.

```python
plot(x,U[:,0],label='$u(x,y=0)$') 
plot(x,U[:,1],label='$u(x,y=1)$')
plot(x,U[:,2],label='$u(x,y=2)$')
plot(x,U[:,3],label='$u(x,y=3)$')
plot(x,U[:,10],label='$u(x,y=10)$')
plot(x,U[:,20],label='$u(x,y=20)$')
plt.xlabel('$x$'); plt.ylabel('$u(x,y)$'); 
plt.title("Evolución de la solución")
xlim([-0.01,1.01]),ylim([-0.30,125.5]),grid(),legend()
show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection="3d")

x = linspace(0,1.0,100)
y = linspace(0,1.0,100)

X,Y = np.meshgrid(x,y)

U = u_xy(x,y)

plt.xlabel('$y$'); plt.ylabel('$x$'); 
plt.title("u(x,y)")
ax.plot_surface(X, Y, U, cmap="hot")
```
